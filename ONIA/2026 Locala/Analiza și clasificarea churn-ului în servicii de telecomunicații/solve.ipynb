```python
import numpy as np
import pandas as pd

X = pd.read_csv("train.csv").drop(columns=["City", "SampleID", "Churn", "Customer ID", "State", "Country", "Zip Code", "Lat Long"])
y = pd.read_csv("train.csv")["Churn"]

X.head()
```

```python
X.info()
```

```python
X.describe()
```

```python
def make_features(X: pd.DataFrame) -> pd.DataFrame:
    X_new = X.copy()
    X_new["FinancialRiskScore"] = 1 * (X["Monthly Charge"] > 70) + 1 * (X["Total Extra Data Charges"] > 10)
    X_new["ServiceQualityScore"] = (X["Avg Speed"] < 50) + 1 * (X["Ping Score"] > 80) + 1 * (X["Link Quality Index"] < 30)
    return X_new
```

```python
X = make_features(X)
```

```python
from catboost import CatBoostClassifier
from sklearn.compose import ColumnTransformer, make_column_selector
from sklearn.discriminant_analysis import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import OneHotEncoder


pipeline = make_pipeline(
    ColumnTransformer(transformers=[
        ('standardscaler', make_pipeline(SimpleImputer(), StandardScaler()), make_column_selector(dtype_exclude=[object])),
        ('onehotencoder', OneHotEncoder(), make_column_selector(dtype_include=[object]))])
)

standardized_X = pipeline.fit_transform(X)

pipeline.get_feature_names_out()
```

```python
model = CatBoostClassifier(iterations=2000, eval_metric="AUC")

model.fit(standardized_X, y)
```

```python
test = pd.read_csv("test.csv").drop(columns=["SampleID", "City", "Customer ID", "State", "Country", "Zip Code", "Lat Long"])
test_idx = pd.read_csv("test.csv")["SampleID"]

test = make_features(test)

standardized_test = pipeline.transform(test)

test_pred = model.predict(standardized_test)
```

```python
answer = pd.concat([
    pd.DataFrame({
        "id": test_idx,
        "subtaskID": 1,
        "answer": test["FinancialRiskScore"]
    }),
    pd.DataFrame({
        "id": test_idx,
        "subtaskID": 2,
        "answer": test["ServiceQualityScore"]
    }),
    pd.DataFrame({
        "id": test_idx,
        "subtaskID": 3,
        "answer": test_pred
    }),
])

answer.to_csv("answer.csv", index=None)
```
