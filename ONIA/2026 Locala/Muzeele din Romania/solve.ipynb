```python
import pandas as pd
import numpy as np

data = pd.read_csv("muzee.csv")

data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
# Task 1 - correct

task_1 = pd.DataFrame({
    "id": 1,
    "subtaskID": 1,
    "answer": 1008
}, index=[1])

task_1
```

```python
# Task 2 - correct

task_2 = pd.DataFrame({
    "id": 2,
    "subtaskID": 2,
    "answer": (data["județul"] == "București").sum()
}, index=[2])

task_2
```

```python
# Task 3 - correct

task_3 = pd.DataFrame({
    "id": 3,
    "subtaskID": 3,
    "answer": 33
}, index=[3])

task_3
```

```python
# Task 4 - correct

by_year = data.groupby(by=data["anul înființării"])
year_with_most_occurances = by_year.size().sort_values().index[-1]

task_4 = pd.DataFrame({
    "id": 4,
    "subtaskID": 4,
    "answer": year_with_most_occurances
}, index=[4])

task_4
```

```python
# Task 5 - correct

by_county_sizes = data.groupby(by=data["județul"]).size()

task_5 = pd.DataFrame({
    "id": by_county_sizes.index,
    "subtaskID": 5,
    "answer": by_county_sizes.values
}, index=None)

task_5
```

```python
# Task 6 - correct

num_of_null_values_series = (data.count(axis="columns") * 100 / data.columns.size).round(2)
task_6 = pd.DataFrame({
    "id": data["_id"],
    "subtaskID": 6,
    "answer": num_of_null_values_series
})

task_6
```

```python
# Task 7 - correct

task_7 = pd.DataFrame({
    "id": 7,
    "subtaskID": 7,
    "answer": (float(num_of_null_values_series.sum()) / float(data["_id"].count())).__round__(2)
}, index=[7])

task_7
```

```python
# Task 8

task_8 = pd.DataFrame({
    "id": 8,
    "subtaskID": 8,
    "answer": round(100 * num_of_null_values_series.map(lambda v: 1 if v == np.max(num_of_null_values_series) else 0).sum() / num_of_null_values_series.values.size, 2)
}, index=[8])

task_8
```

```python
answer = pd.concat(
    [task_1, task_2, task_3, task_4, task_5, task_6, task_7, task_8]
)
answer.to_csv("answer.csv", index=None)
```
